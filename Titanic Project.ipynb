```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
df_1=pd.read_csv('test.csv')
df.head(10)
df_1.head()
```

```python
df['surname'] = df['Name'].apply(lambda x: x.split(' ')[1])
name_counts= df['surname'].value_counts()
df['sur']=df['surname'].apply(lambda x : 'other' if name_counts[x]<10 else x)
df.drop(['surname'],axis='columns')
```

```python
df_1['surname'] = df_1['Name'].apply(lambda x: x.split(' ')[1])
name_counts_1= df_1['surname'].value_counts()
df_1['sur']=df_1['surname'].apply(lambda x : 'other' if name_counts_1[x]<10 else x)
df_1.drop(['surname'],axis='columns')
```

```python
df2= df.drop(['Name','Ticket','Cabin','Fare','SibSp','Parch','PassengerId'],axis='columns')
df_2= df_1.drop(['Name','Ticket','Cabin','Fare','SibSp','Parch'],axis='columns')

```

```python
df3=df2.dropna()
df_3=df_2
PI=df_3.PassengerId
df_3= df_3.drop(['PassengerId'],axis='columns')
```

```python
df4 = pd.get_dummies(df3,columns=['Sex','Embarked','sur'])
df_4 = pd.get_dummies(df_3,columns=['Sex','Embarked','sur'])

df4= df4.drop(['sur_other','surname'],axis='columns')
df_4= df_4.drop(['sur_other','surname'],axis='columns')
df_4
```

```python
x=df_4.mode()
x
df_4.fillna(x,inplace=True)
df_4
```

```python
X_train=df4.drop(['Survived'],axis='columns')
y_train=df4.Survived
X_test=df_4
X_test
```

```python
from xgboost import XGBClassifier
xgb= XGBClassifier()
xgb.fit(X_train,y_train)
y= xgb.predict(X_test)
y
```

```python
X_test['Survived']=y
X_test
```

```python
Y=pd.concat([X_test,PI],axis='columns')
Y
```

```python
Y=Y.drop(['Pclass','Age','Sex_female','Sex_male','Embarked_C','Embarked_Q','Embarked_S'],axis='columns')
cols= ['PassengerId','Survived']
Y=Y[cols]
Y
```

```python
Y.to_csv('titanic_prediction.csv',index=False)
```
